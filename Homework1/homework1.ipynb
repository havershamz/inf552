## 1. (a)
- ### Download compressed dataset archive from the website and decompress it into the same directory where homework notebook is.

- ### It is a binary classification problem, so *column_2C* is needed. And I don't use WEKA for now, so use *column_2C.dat* as data sourse. 

- ### Open the data in text editor, there is no header row and every individual lists in each row, with 6 predictors and a indicator. contents are seperated by space. 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.DataFrame(pd.read_table('vertebral_column_data/column_2C.dat',header=None,sep='\s+'))
df_clean = df.copy()
df.columns =['pelvic incidence', 'pelvic tilt', 'lumbar lordosis angle',
'sacral slope', 'pelvic radius', 'grade of spondylolisthesis', 'outcome']
df_clean[6] = df['outcome'].map(lambda x:1 if x == 'AB' else 0)

df_clean # show the clean data frame in a intuitive way.
```

## 1. (b) i. 
### Plot scatter plot with pairs of predictors.
### There are 6 predictors, meaning 6x6 subplots.

```python
colors = df_clean[6].map(lambda x:'pink' if x == 1 else 'lightblue')
plt.figure(figsize = (36, 36), dpi = 98)

i = 0
while (i < 36):
    x1 = i//6
    x2 = i%6
    plt.subplot(6, 6, i+1)
    plt.scatter(df_clean[x1], df_clean[x2], c = colors)
    plt.xlabel(df.columns[x1])
    plt.ylabel(df.columns[x2])
    i = i+1

```

## 1. (b) ii. 
### Plot boxplot of each variable with class0 vs class1

```python
cls1 = df_clean[df_clean[6] == 1] # class 1
cls0 = df_clean[df_clean[6] == 0] # class 0

fig, axes = plt.subplots(nrows=1, ncols=6, figsize=(36, 6), dpi = 98)

box_plot =[]
for i in range(0,6):
 box_plot.append(axes[i].boxplot([cls0[i],cls1[i]], labels=['class0', 'class1'], vert=True, patch_artist=True, widths = 0.5))   

colors = ['lightblue', 'pink']
for bplot in box_plot:
    for patch, color in zip(bplot['boxes'], colors):
        patch.set_facecolor(color)
i = 0
for ax in axes:
    ax.set_xticks([y+1 for y in range(len([cls0[0],cls1[0]]), )])
    ax.set_xlabel(df.columns[i])
    i = i+1
    
plt.show()
```

## 1. (b) iii. 
### Divide data into training set and test set.

```python
df_tr = pd.concat([cls1[0:140], cls0[0:70]])
df_ts = pd.concat([cls1[140:], cls0[70:]])
df_ts
```

## 1. (c) i. 
### Homework document provides a pretty solid machine learning package "sklearn". It contains a module for KNN classifier. I use it for the following questions.

```python
from sklearn.neighbors import KNeighborsClassifier
```

## 1. (c) ii.
### First, compute the best k with minimum test error rate.

```python
tr_err_list = [] # create a list of training error rate list for each k.
ts_err_list = [] # create test error rate list.

knc_euc = KNeighborsClassifier(n_neighbors = 0) # create k nearest nerghbors classifier with Euclidean Distance.
for k in range(208,0,-3):
    knc_euc.set_params(n_neighbors = k) # change parameter k.
    knc_euc.fit(df_tr.loc[:,:5], df_tr[6]) # train the model with training set.
    tr_err = 1-knc_euc.score(df_tr.loc[:,:5], df_tr[6]) # compute the training error rate.
    ts_err = 1-knc_euc.score(df_ts.loc[:,:5], df_ts[6]) # compute the test error rate.
    tr_err_list.append(tr_err)
    ts_err_list.append(ts_err)
    
plt.gca().invert_xaxis()
plt.ylabel("Error Rate")
plt.xlabel("K Value")
plt.title('Error Rate VS K Value')

plt.plot(range(208,0,-3),tr_err_list, color='lightblue', label = 'Training Set Error') 
plt.plot(range(208,0,-3),ts_err_list, color='pink', label = 'Test Error') 

ts_err_min_k = 208-3*ts_err_list.index(min(ts_err_list)) # calculate the k with the minimum test error rate.
print("Minimum test error rate is {}, k is {}.".format(round(min(ts_err_list), 2), ts_err_min_k))
```

###   With k = 4, compute the confusion matrix, true positive rate, true negative rate, precision and F-score.

```python
from sklearn.metrics import confusion_matrix, classification_report
import itertools

"""
This is a function I found on sklearn document and I made a little tweak.
Original URL shows below.
http://scikit-learn.org/stable/auto_examples/model_selection/plot_confusion_matrix.html
"""
def plot_confusion_matrix(cm, classes, title='Confusion matrix', cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    """
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = range(len(classes))
    plt.xticks(tick_marks, classes)
    plt.yticks(tick_marks, classes)

    fmt = 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    
knc_euc = KNeighborsClassifier(n_neighbors = 4)
knc_euc.fit(df_tr.loc[:,:5], df_tr[6])

true_value = df_ts[6]
predict_value = knc_euc.predict(df_ts.loc[:,:5])
cnf_matrix = confusion_matrix(true_value, predict_value,labels = [0,1]) # compute confusion matrix.

plot_confusion_matrix(cnf_matrix,classes = ['NORMAL','ABNORMAL'],title='confusion matrix') # plot confusion matrix.

tn, fp, fn, tp = cnf_matrix.ravel()
true_pos_rate = tp/(tp + fn) # calculate true positive rate.
ture_neg_rate = tn/(tn + fp) # calculate true negtive rate.
print('The true positive rate is {}, the true negtive rate is {}.'.format(round(true_pos_rate, 2), round(ture_neg_rate, 2)))
print(classification_report(true_value, predict_value, target_names=['NORMAL','ABNORMAL'])) # the report shows precision and F-score
```

## 1. (c) iii.
### Plot the  *best test error rate*, against the size of training set.

```python

def test_error_rate_list(training_set_size, k_range, knc):
    """
    This function takes in training set at different size, and return
    the test error rate list reagarding the range of k.
    Variable k_range takes in the range of k.
    Variable knc takes in the KNeighborsClassifier model.
    """
    ts_err_list = [] # create test error list.

    for k in k_range:
        knc.set_params(n_neighbors = k) # change parameter k.
        
        df_training = pd.concat([cls0[0:training_set_size//3], cls1[0:(training_set_size-training_set_size//3)]])
        knc.fit(df_training.loc[:, :5], df_training[6]) # train the model with training set.  
        
        ts_err = 1-knc.score(df_ts.loc[:, :5], df_ts[6]) # compute the test error.
        ts_err_list.append(ts_err)
    return ts_err_list

knc_euc = KNeighborsClassifier(n_neighbors = 0) # Euclidean Distance
best_test_error_list = [] # create a list of the best test error rate regarding training set size. 
for n in range(10,211,10):   
    best_test_error_list.append(min(test_error_rate_list(n, range(1,n,5), knc_euc)))
 
plt.title('Learning Curve')
plt.xlabel('Training Set Size')
plt.ylabel('Minimum Test Error')
plt.plot(range(10,211,10), best_test_error_list)
plt.show()
```

## 1. (d) i. A.
### Use Manhattan Distance, compute test error rate with different k.

```python
knc_mht = KNeighborsClassifier(n_neighbors = 0, p = 1) #  Manhattan Distance

ts_err_ls_mht = test_error_rate_list(210, (range(1,200,5)), knc_mht) 
min_ts_err_mht = min(ts_err_ls_mht) # compute minimum test error rate.
k_mht = range(1,200,5)[ts_err_ls_mht.index(min_ts_err_mht)] # compute ragarding k.

print("Use Manhattan Distance, the minimum test error rate is {}, k is {}.".format(round(min_ts_err_mht, 2), k_mht))
plt.xlabel('K Value')
plt.ylabel('Test Error Rate')
plt.plot(range(1,200,5), ts_err_ls_mht)
plt.gca().invert_xaxis()
plt.show()

print('K Value   Test Error Rate')
for i in range(len(ts_err_ls_mht)):
    print("{:<3}\t  {:<3}".format(1+5*i,round(ts_err_ls_mht[i], 2))) 
```

## 1. (d) i. B.
### Calculate the best lg(p).

```python
best_lgp = 0.1
best_ts_err = 1
for i in range(1, 11, 1):
    p = 10**(i*0.1)
    knc_min = KNeighborsClassifier(n_neighbors = 6, p = p) # Minkowski metric with k = 6.
    knc_min.fit(df_tr.loc[:,:5],df_tr[6])
    ts_err = 1-knc_min.score(df_ts.loc[:,:5],df_ts[6]) # compute the testing error rate.
    if ts_err < best_ts_err:
        best_lgp = i*0.1
        best_ts_err = ts_err

print('The best lg(p) is ', round(best_lgp,2))
print("The minimum test error rate is {}, k is {}.".format(round(best_ts_err, 2), 6))
```

## 1. (d) i. C.
### Use Chebyshev Distance, compute test error rate with different k.

```python
knc_cbs = KNeighborsClassifier(n_neighbors = 0, metric = 'chebyshev') # Chebyshev Distance

ts_err_ls_cbs = test_error_rate_list(210, (range(1,200,5)), knc_cbs) 

min_ts_err_cbs = min(ts_err_ls_cbs)
k_cbs = range(1,200,5)[ts_err_ls_cbs.index(min(ts_err_ls_cbs))]

print("Use Chebyshev Distance, the minimum test error rate is {}, k is {}.".format(round(min_ts_err_cbs, 2), k_cbs))
plt.xlabel('K Value')
plt.ylabel('Test Error Rate')
plt.plot(range(1,200,5), ts_err_ls_cbs)
plt.gca().invert_xaxis()
plt.show()

print('K Value   Test Error Rate')
for i in range(len(ts_err_ls_cbs)):
    print("{:<3}\t  {:<3}".format(1+5*i,round(ts_err_ls_cbs[i], 2))) 
```

## 1. (d) ii.
### Use Mahalanobis Distance, compute test error rate with different k.

```python
knc_mhl = KNeighborsClassifier(n_neighbors = 0, metric = 'mahalanobis', algorithm = 'brute', metric_params = {'V': np.cov(df_tr.loc[:,0:5])}) # Mahalanobis Distance.

ts_err_ls_mhl = test_error_rate_list(210, (range(1,200,5)), knc_mhl) 

min_ts_err_mhl = min(ts_err_ls_mhl)
k_mhl = range(1,200,5)[ts_err_ls_mhl.index(min_ts_err_mhl)]

print("Use Mahalanobis Distance, the minimum test error rate is {}, k is {}. ".format(round(min_ts_err_mhl, 2), k_mhl))
plt.xlabel('K Value')
plt.ylabel('Test Error Rate')
plt.plot(range(1,200,5), ts_err_ls_mhl)
plt.gca().invert_xaxis()
plt.show()

print('K Value   Test Error Rate')
for i in range(len(ts_err_ls_mhl)):
    print("{:<3}\t  {:<3}".format(1+5*i,round(ts_err_ls_mhl[i], 2))) 
```

## 1. (e)
###  Use weighted voting to compute best test error rate with different distance metrics.

```python
knc_euc_weighted = KNeighborsClassifier(n_neighbors = 0, weights = 'distance')
ts_err_ls_euc_w = test_error_rate_list(210, (range(1,200,5)), knc_euc_weighted)
min_ts_err_euc_w = min(ts_err_ls_euc_w)
k_euc_w = range(1,200,5)[ts_err_ls_euc_w.index(min_ts_err_euc_w)]
print("When weighted with distance, use Euclidean Distance, the minimum test error rate is {}, k is {}.".format(round(min_ts_err_euc_w, 2), k_euc_w))

knc_mht_weighted = KNeighborsClassifier(n_neighbors = 0, p = 1, weights = 'distance')
ts_err_ls_mht_w = test_error_rate_list(210, (range(1,200,5)), knc_mht_weighted)
min_ts_err_mht_w = min(ts_err_ls_mht_w)
k_mht_w = range(1,200,5)[ts_err_ls_mht_w.index(min_ts_err_mht_w)]
print("When weighted with distance, use Manhattan Distance, the minimum test error rate is {}, k is {}.".format(round(min_ts_err_mht_w, 2), k_mht_w))

knc_cbs_weighted = KNeighborsClassifier(n_neighbors = 0, metric = 'chebyshev', weights = 'distance')
ts_err_ls_cbs_w = test_error_rate_list(210, (range(1,200,5)), knc_cbs_weighted)
min_ts_err_cbs_w = min(ts_err_ls_cbs_w)
k_cbs_w = range(1,200,5)[ts_err_ls_cbs_w.index(min_ts_err_cbs_w)]
print("When weighted with distance, use Chebyshev Distance, the minimum test error rate is {}, k is {}.".format(round(min_ts_err_cbs_w, 2), k_cbs_w))
```
